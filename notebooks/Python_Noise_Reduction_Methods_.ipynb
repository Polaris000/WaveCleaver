# De-Noise with Spectral Gating

Spectral gating is a form of noise reduction that takes the spectrogram of an audio input and calculates the mean and standard deviation over each frequency band. A threshold for each band is determined based on some number of standard deviations away from the mean (z-score). This threshold is the "gate", and any signal outside of the threshold for this frequency is considered noise and filterd out using masks of only the desired signals.

[noisereduce](https://github.com/timsainb/noisereduce?tab=readme-ov-file)
